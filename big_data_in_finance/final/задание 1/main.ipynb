```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('RC_F02_12_2018_T02_12_2023.xlsx')
```

```python
df['last_day'] = pd.to_datetime(df['data']).dt.to_period('M').dt.to_timestamp('M') + pd.offsets.MonthEnd(0)
```

```python
df_filtered = df.groupby('last_day')['curs'].last().reset_index()
```

```python
data_values = df_filtered['last_day'].tolist()
curs_values = df_filtered['curs'].tolist()
```

```python
plt.plot(data_values, curs_values)
plt.xlabel('Дата')
plt.ylabel('Курс')
plt.title('График курса доллара')
plt.xticks(rotation=45)
plt.show()
```

```python
# Чтение файла CSV

data = pd.read_csv('russia.index.cpi (statbureau.org).csv', index_col=0)

# Получение списка дат
dates = []
inflation_values = []

for year in data.index:
    for month in data.columns:
        if (year == 2018 and month == 'Декабрь') or year > 2018:
            date = f'{month} {str(year)}'  # Форматирование даты в строку
            dates.append(date)
            inflation_values.append(data.loc[year, month])
# Построение графика
plt.plot(dates, inflation_values)
plt.xticks(rotation=45)
plt.xlabel('Дата')
plt.ylabel('Инфляция')
plt.title('Инфляция по месяцам')
plt.show()
```

Закономерность между курсом рубля и инфляцией не является однозначной. Однако, стоит отметить, что резкие ослабления курса национальной валюты приводили к скачкам инфляции. Это можно увидеть на графиках в нескольких местах, например в феврале 2022 года. Это происходило отчасти из-за того, что при ослаблении курса рубля повышались цены на импортные товары, является проинфляционным фактором, а также повышало инфляционные ожидания населения и бизнеса, что также в свою очередь предполагает закономерное увеличение цен на товары и услуги.
