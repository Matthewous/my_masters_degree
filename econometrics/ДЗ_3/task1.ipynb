```python
# Import libraries
import statsmodels.api as sm
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
```

```python
df_oil = pd.read_excel('brent.xlsx')
```

```python
df_oil['Date'] = pd.to_datetime(df_oil['Дата'])
df_oil['Значение'] = pd.to_numeric(df_oil['Значение'])
```

```python
# Построение графика 
plt.plot(df_oil['Date'], df_oil['Значение'])
plt.xticks(rotation=90)
plt.xlabel('Дата')
plt.ylabel('Цена')
plt.title('Временной ряд цен')
plt.show()
```

```python
# Проверка стационарности с помощью теста Дики-Фуллера
result = adfuller(df_oil['Значение'])
print('ADF статистика:', result[0])
print('p-value:', result[1])
```

0.17 > 0.05 => ряд нестационарен
