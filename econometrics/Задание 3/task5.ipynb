```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import f_regression
import statsmodels.api as sm
```

```python
df_oil = pd.read_excel('/Users/matthew/Documents/GitHub/my_masters_degree/econometrics/Задание 3/brent.xlsx')
```

```python
df_oil['Date'] = pd.to_datetime(df_oil['Дата'])
df_oil['Значение'] = pd.to_numeric(df_oil['Значение'])
```

```python
# Преобразование временного ряда в лог-приросты цен

df_oil['log_returns'] = np.log(df_oil['Значение']).diff()
```

```python
df_oil.dropna()
```

```python
df_oil['log_returns'].plot()
```

```python
sm.graphics.tsa.plot_acf(df_oil['log_returns'], lags=1)
```
